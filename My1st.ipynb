```python
def MAE(y_test, y_predict):
    return np.sum(abs(y_test-y_predict))/len(y_test)

def MSE(y_test, y_predict):
    return np.sum((y_test - y_predict)**2) / len(y_test)

def RMSE(y_test, y_predict):
    return sqrt(MSE(y_test, y_predict))

def MAPE(y_test, y_predict):
    return np.mean(abs((y_test - y_predict) / y_test))

def R2(y_test, y_predict):
    return 1 - (MSE(y_test, y_predict) / (np.sum((y_test - y_test.mean())**2) / len(y_test)))
```

```python

class RidgeLinearRegression:
    def __init__(self, alpha=1.0):
        self.alpha = alpha

    def fit(self, X, y):
        m, n = X.shape
        X_b = np.c_[np.ones((m, 1)), X]

        # calculate the normal equations with ridge regularization
        A = np.eye(n+1)
        A[0, 0] = 0
        A[1:, 1:] = self.alpha * np.identity(n)
        self.theta = np.linalg.inv(X_b.T.dot(X_b) + A).dot(X_b.T).dot(y)

    def predict(self, X):
        m, n = X.shape
        X_b = np.c_[np.ones((m, 1)), X]
        return X_b.dot(self.theta)
```
